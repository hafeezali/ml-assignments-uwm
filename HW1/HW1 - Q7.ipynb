```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Solution 1

mean = [1,-1]
cov = [[2,0],[0,2]]
```

```python
x, y = np.random.multivariate_normal(mean, cov, 500).T
```

```python
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

```python
# Solution 2
mean1 = [5,0]
cov1 = [[1,0.25],[0.25,1]]
mean2 = [-5,0]
cov2 = [[1,-0.25],[-0.25,1]]
```

```python
x, y = 0.3 * np.random.multivariate_normal(mean1, cov1, 100).T + 0.7 * np.random.multivariate_normal(mean2, cov2, 100).T
```

```python
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```
